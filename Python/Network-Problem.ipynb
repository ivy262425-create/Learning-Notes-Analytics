```python
import networkx as nx
#VARIABLE NAME
left = ['A', 'C', 'F', 'H','I','K'] # len = 6
right = ['B', 'D','E','G','J','L'] # len = 6
#Dictionary
edges = [
    ('A','E'), ('F','E'), ('I','E'), ('I','J'), ('F','J'),
    ('F','G'), ('C','G'), ('H','G'), ('K','G'), ('H','D'), ('H','L'), ('C','D')
]

# Define the Source and Sink nodes
SOURCE = 's'
SINK = 't'

### Construct the graph ###
G = nx.DiGraph()

# 1. Add Source and Sink nodes with total supply/demand
G.add_node(SOURCE, demand=-1)  # demand = -6 (Total Supply)
G.add_node(SINK, demand=1)    # demand = 6 (Total Demand)

# 2. Add all original nodes with zero demand
for node in left + right:
    G.add_node(node, demand=0)

# 3. Add edges from SOURCE to all LEFT nodes (capacity 1, weight 0)
for u in left:
    G.add_edge(SOURCE, u, capacity=1, weight=0)

# 4. Add edges from all RIGHT nodes to SINK (capacity 1, weight 0)
for v in right:
    G.add_edge(v, SINK, capacity=1, weight=0)

# 5. Add edges for the actual matching problem (capacity 1, weight -1)
for u, v in edges:
    G.add_edge(u, v, capacity=1, weight=-1)

####
flowCost, flowDict = nx.network_simplex(G)
print(f"optimal flow cost: {flowCost}")
```
